# **Tuple Duplication**

Tuple duplication are the important issues in data integration during data mining Duplicate tuples and attributes cause inconsistencies in attributes and inconsistencies in the database or data sets.  Duplicate tuples generally occur due to inaccurate data entry or updating the files of similar data occurrences.

Removal of duplicate tuples is considered as the primary step in data processing in all applications. Before performing the operations on the data set or developing the models data set is cleaned to remove the noise data. Duplicate tuples are the noise data that affect the accuracy of the models.

Creating the dataset

```python
dict1 = {'brand':['Yum Yum','Yum Yum','Indomine','Hello','Indomine','Indomine'],
          'style':['Cup','Cup','Pack','Cup','Hub','Pack'],
           'ratings':[4.0,4.0,2.6,1.8,3.5,4.4]}

df1 = pd.DataFrame(dict1)
df1
```

Finding out duplicates.

```python
duplicates=df1[df1.duplicated()]
print(duplicates)
```

Finding out duplicates in a coloumn

```python
dupli_style=df1[df1.duplicated('style')]
print(dupli_style)
```

Removing duplicates

```python
df1.drop_duplicates()
```

Removing duplicates in a specific column

```python
df1.drop_duplicates(subset=['brand'])
```

 When you set keep to 'last', it means that only the last occurrence of each duplicated row, based on the specified subset of columns, will be retained, and all previous occurrences will be removed.

```python
df1.drop_duplicates(subset=['brand', 'style'], keep='last')
```

'first': Keep the first occurrence of each duplicated row and remove subsequent occurrences.

```python
df1.drop_duplicates(subset=['brand', 'style'], keep='first')
```

False or None: Remove all duplicated rows, keeping none.

```python
df1.drop_duplicates(subset=['brand', 'style'], keep=False)
```

inplace=False: If you set inplace=False the operation will return a new object with the changes applied, leaving the original object unchanged.

```python
df1.drop_duplicates(subset=['brand'],inplace=False)
```

inplace=True: If you set inplace=True, the operation will modify the original object directly, and it won't return a new object. This means that the changes will be made to the existing DataFrame or Series,

```python
df1.drop_duplicates(subset=['brand'],inplace=True)
df1
```
