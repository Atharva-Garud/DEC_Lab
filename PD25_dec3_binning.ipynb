# **Data Smoothening**

Data smoothing and binning methods are techniques used to preprocess and transform data, particularly in data analysis and data mining, to make it more manageable and suitable for further analysis. These methods help reduce noise, discover patterns, and simplify complex datasets.



**Bin Smoothing**:



* Bin smoothing involves grouping data points into bins or intervals, then
  replacing each data point within a bin with a representative value
* This method simplifies the data while preserving general trends and  
  patterns. It is often used in histogram construction.


Binning Dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'values': [0, 3, 8, 10, 12, 20, 25, 30, 40, 43, 53, 58, 60]}
df = pd.DataFrame(data)

bins = [0, 20, 40, 60]
df['binned'] = pd.cut(df['values'], bins)

df['binned'].value_counts().plot(kind='bar', title='Binned Data - DataFrame')
plt.xlabel('Bins')
plt.ylabel('Count')
plt.show()
```

Binning CSV

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('sample-data.csv')

bins = [10,20,30,40,50]
df['binned'] = pd.cut(df['Age'], bins)

df['binned'].value_counts().plot(kind='bar', title='Binned Data - CSV')
plt.xlabel('Bins')
plt.ylabel('Count')
plt.show()
```

Binning Inbuilt function

```python
import numpy as np

import matplotlib.pyplot as plt


data = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 58, 60]


bins = [0, 25, 50, 75]
hist, bin_edges = np.histogram(data, bins=bins)


plt.bar(bin_edges[:-1], hist, width=20, align='edge')
plt.title('Binned Data - Inbuilt Function')
plt.xlabel('Bins')
plt.ylabel('Count')
plt.show()
```

Binning User defined function

```python
import numpy as np
import matplotlib.pyplot as plt

data = [10, 15, 20, 25, 30, 35, 40, 45, 50, 52, 55, 58, 60]


bins = [0, 20, 40, 60]


plt.hist(data, bins=bins, edgecolor='k')
plt.title('Binned Data - Custom Function')
plt.xlabel('Bins')
plt.ylabel('Count')
plt.show()
```
