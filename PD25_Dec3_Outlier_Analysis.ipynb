# **Outlier analysis**

Outlier analysis, also known as outlier detection or anomaly detection, is a crucial step in data mining and data analysis. It involves identifying and handling data points that are significantly different from the majority of the data.



*   Outliers are data points that are unusually distant from other data
    points in a dataset.
*  They can be exceptionally high or low values, or they can have unusual
   patterns or characteristics compared to the rest of the data.
*  Outliers can be genuine data anomalies or errors, and distinguishing between
   the two is an important aspect of outlier analysis.









```python
import pandas as pd
import matplotlib.pyplot as plt

import numpy as np

dataset= [11,10,12,14,12,15,14,13,15,102,12,14,17,19,107, 10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]

df = pd.DataFrame(dataset)
print(df)
```

Finding mean and S.D.

```python
mean = np.mean(dataset)
std = np.std(dataset)
print('mean of the dataset is', mean)
print('std. deviation is', std)
```

Finding outliers using z-score

```python

threshold = 3
outlier = []
for i in dataset:
    z = (i-mean)/std
    if np.abs(z) > threshold:
        outlier.append(i)
print('outlier in dataset is', outlier)
```

Finding IQR and quartiles

```python
quartiles = df[0].quantile([0.25, 0.75])
iqr = quartiles[0.75] - quartiles[0.25]
q1=quartiles[0.25]
q2=quartiles[0.75]
print("Q1: " ,q1)
print("Q2: " ,q2)
print("IQR: " ,iqr)
```

Upper and lower bound

```python
lower_bnd=q1-(1.5*iqr)
upper=q2-(1.5*iqr)

print("Lower bound is: ",lower_bnd)
print("Upper bound is: ",upper)
```
