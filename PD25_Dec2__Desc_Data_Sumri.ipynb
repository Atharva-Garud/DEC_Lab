```python
import pandas as pd
import numpy as np

df=pd.read_csv('sample_data.csv')
pd.read_csv('sample_data.csv')
```

```python
col1 = df['Age']
col2 = df['Salary']
```

```python
mean1 = col1.mean()
median1 = col1.median()
mode1 = col1.mode().iloc[0]

print("Measures of Central Tendency for Age:")
print("Mean:", mean1)
print("Median:", median1)
print("Mode:", mode1)
```

```python
mean2 = col2.mean()
median2 = col2.median()
mode2 = col2.mode().iloc[0]

print("\nMeasures of Central Tendency for Salary:")
print("Mean:", mean2)
print("Median:", median2)
print("Mode:", mode2)
```

```python
range1 = col1.max() - col1.min()
iqr1 = col1.quantile(0.75) - col1.quantile(0.25)

print("Range:", range1)
print("Interquartile Range (IQR):", iqr1)
print("Q1: ",col1.quantile(0.25))
print("Q2: ",col1.quantile(0.50))
print("Q3: ",col1.quantile(0.75))
```

```python
range2 = col2.max() - col2.min()
iqr2 = col2.quantile(0.75) - col2.quantile(0.25)

print("Range:", range2)
print("Interquartile Range (IQR):", iqr2)
print("Q1: ",col2.quantile(0.25))
print("Q2: ",col2.quantile(0.50))
print("Q3: ",col2.quantile(0.75))
```
